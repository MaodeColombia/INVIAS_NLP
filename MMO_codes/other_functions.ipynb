## POR IMPLEMENTAR: Recorre todos los archivos JSON en una carpeta específica y verifica si la clave 'content' contiene un patrón predefinido. Si un archivo no cumple con la condición, lo reporta.

````python
import os
import json
import re

# Expresión regular para detectar el patrón específico en los documentos JSON
pattern = re.compile(
    r"_{80,}\nInstituto Nacional de Vías\nPBX: \(\+57\) 601 377 0600 Línea gratuita: 018000117844\n"
    r"http://www\.invias\.gov\.co\nCorreo institucional: atencionciudadano@invias\.gov\.co\n"
    r"\d+/\d+\nFecha: \d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}\n"
    r"Radicado: \d{4}[A-Z]-[A-Z]+-\d+\nDestino: .*\nAnexos: \d+"
)

# Ruta donde se encuentran los archivos JSON
folder_path = "../assets/DG_docs/jsonembedding_pdf/"  # MODIFICA con la ubicación de los archivos JSON

# Diccionario para almacenar los resultados de la búsqueda
pattern_found = {}

# Recorrer cada archivo JSON en la carpeta
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):  # Filtrar solo archivos con extensión .json
        file_path = os.path.join(folder_path, filename)  # Construir la ruta completa del archivo

        try:
            # Intentar abrir el archivo y cargar su contenido como JSON
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)

            found = False  # Bandera para indicar si se encuentra el patrón en el archivo

            # Iterar sobre la lista de textos dentro del JSON
            for entry in data.get("text", []):  
                content = entry.get("content", "")  # Extraer el contenido del texto
                if pattern.search(content):  # Buscar el patrón en el contenido
                    found = True
                    break  # Si se encuentra el patrón, detener la búsqueda en este archivo

            # Almacenar el resultado en el diccionario
            pattern_found[filename] = found

        except Exception as e:
            # Si ocurre un error al procesar el archivo, almacenarlo en el diccionario
            pattern_found[filename] = f"Error al procesar: {e}"

# Imprimir los resultados de la búsqueda
for file, found in pattern_found.items():
    print(f"{file}: {'Patrón encontrado' if found else 'Patrón NO encontrado'}")
````

## BAJO NINGUNA CIRCUNSTANCIA BORRAR:  este codigo valida que los JSON estan bien

````python
import os
import json

# Configuración
JSON_FOLDER = "../assets/DG_docs/jsonembedding_pdf"  # Ruta donde están los JSONs
EXPECTED_DIMENSION = 1536  # Dimensión esperada para text-embedding-3-small

def validate_json(file_path):
    """ Valida un archivo JSON y revisa si los embeddings tienen la dimensión correcta. """
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)
        
        # Verificar que las claves principales existen
        if "pdf_filename" not in data or "text" not in data:
            return f"❌ ERROR: Falta 'pdf_filename' o 'text' en {file_path}"

        if not isinstance(data["text"], list):
            return f"❌ ERROR: 'text' debe ser una lista en {file_path}"

        # Revisar cada página del documento
        for page_data in data["text"]:
            # Verificar que cada página tenga las claves necesarias
            if not all(key in page_data for key in ["page", "content", "OpenAI_embedding"]):
                return f"❌ ERROR: Falta 'page', 'content' o 'OpenAI_embedding' en {file_path} (Página {page_data.get('page', 'Desconocida')})"

            # Obtener el embedding
            embedding = page_data["OpenAI_embedding"]

            # Validar si el embedding es 'None'
            if embedding is None:
                return f"❌ ERROR: Embedding es 'None' en {file_path} (Página {page_data['page']})"

            # Validar que el embedding sea una lista y tenga la dimensión correcta
            elif not isinstance(embedding, list) or len(embedding) != EXPECTED_DIMENSION:
                return f"❌ ERROR: Embedding de tamaño incorrecto en {file_path} (Página {page_data['page']}, esperado {EXPECTED_DIMENSION}, obtenido {len(embedding)})"

        return f"✅ OK: {file_path} está bien estructurado"

    except json.JSONDecodeError:
        return f"❌ ERROR: JSON mal formado en {file_path}"
    except Exception as e:
        return f"❌ ERROR inesperado en {file_path}: {str(e)}"

def validate_all_jsons(json_folder):
    """ Valida todos los archivos JSON en una carpeta. """
    files = [f for f in os.listdir(json_folder) if f.endswith(".json")]
    if not files:
        print("⚠ No se encontraron archivos JSON en la carpeta.")
        return

    for file in files:
        file_path = os.path.join(json_folder, file)
        print(validate_json(file_path))

# Ejecutar validación
validate_all_jsons(JSON_FOLDER)
````
